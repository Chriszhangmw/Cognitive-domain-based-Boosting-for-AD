```python
from PIL import ImageEnhance
from PIL import Image
import imageio as iio
import scipy.ndimage as ndi
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = Image.open('./1/122.png')
# image.show()
#
enh_bri = ImageEnhance.Brightness(image)
brightness = 4
image_enhanced = enh_bri.enhance(brightness)
# image_enhanced.show()
#
enh_col = ImageEnhance.Color(image)
color = 4
image_colored = enh_col.enhance(color)
# image_colored.show()
# 
enh_con = ImageEnhance.Contrast(image)
contrast = 4
image_contrasted = enh_con.enhance(contrast)
# image_contrasted.show()
plt.imshow(image_contrasted, cmap='bone')
plt.axis('off')
plt.show()
```

```python
image = Image.open('./1/122.png')
# 
enh_con = ImageEnhance.Contrast(image)
contrast = 4
image_contrasted = enh_con.enhance(contrast)
print(image_contrasted.convert('L'))
image_array = np.array(image_contrasted)
t = 128
image_array[image_array>t]=255
image_array[image_array<t]=0
plt.imshow(image_array,cmap='Reds')
plt.show()
```
